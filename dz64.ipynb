## Завдання:

За допомогою графічного редактора визначте точки відповідності (кути дошки) на зображеннях **chess_1 та chess_2** , що додані до репозиторію лекції. 

Проведіть гомографічне перетворення для приведення зображення **chess_2** до вигляду базового зображення **chess_1.**

## Розв'язок

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
im_src = cv2.imread('chess_2.png')
pts_src = np.array([[220, 315], [217, 69], [380, 348],[470, 115]])
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.axis('off')
plt.show();
```

```python
im_dst = cv2.imread('chess_1.png')
pts_dst = np.array([[210, 80],[447, 71],[214, 219],[441, 229]])
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.axis('off')
plt.show();
```

```python
h, status = cv2.findHomography(pts_src, 
                               pts_dst,
                               )
```

```python
im_out = cv2.warpPerspective(im_src, h,
                             (im_dst.shape[1],
                              im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.axis('off')
plt.show();
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 15), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(im_dst, cmap=plt.cm.gray)
ax1.set_title('chess_1')
ax2.axis('off')
ax2.imshow(im_src, cmap=plt.cm.gray)
ax2.set_title('chess_2')
ax3.axis('off')
ax3.imshow(im_out, cmap=plt.cm.gray)
ax3.set_title('chess_2 to chess_1')
plt.show()
```
