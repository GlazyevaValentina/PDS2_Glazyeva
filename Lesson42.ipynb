```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
df = datasets.load_wine()
model = KMeans(n_clusters=4)
model.fit(df.data)
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(df.data)
```

```python
predictions = model.predict(df.data)
```

```python
print(predictions)
```

```python
cluster_0 = df.data[predictions == 0]
cluster_1 = df.data[predictions == 1]
cluster_2 = df.data[predictions == 2]
cluster_3 = df.data[predictions == 3]
```

```python
plt.scatter(cluster_0[:,0] , cluster_0[:,1] , color = 'red')
plt.scatter(cluster_1[:,0] , cluster_1[:,1] , color = 'black')
plt.scatter(cluster_2[:,0] , cluster_2[:,1] , color = 'yellow')
plt.scatter(cluster_3[:,0] , cluster_3[:,1] , color = 'green')
```

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
dbscan = DBSCAN(eps=0.4, min_samples=4)
```

```python
dbscan.fit(df.data)
```

```python
pca = PCA(n_components=2).fit(df.data)
pca_2d = pca.transform(df.data)
```

```python
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='y', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='d')
    elif dbscan.labels_[i] == 2:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='o')
```
